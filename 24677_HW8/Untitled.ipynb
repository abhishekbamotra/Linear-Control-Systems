```python
from mpl_toolkits import mplot3d
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt


fig = plt.figure()
x1 = np.linspace(-5,5,100)
x2 = x1

X, Y = np.meshgrid(x1,x2)

U = Y
V = 0*Y

plt.streamplot(X, Y, U, V, density = 0.5, color = 'red', linewidth = 2)
```

```python
fig = plt.figure()

ax = plt.axes(projection='3d')
space = np.linspace(-15, 15, 2000)
x1 = space
x2 = space
X,Y = np.meshgrid(x1,x2)
vdot = -4 * X**4 * Y**2
ax.plot_surface(X, Y, vdot)
```
